```python
#importing libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# load dataset to pandas dataframe
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
# explore dataset
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
# information about the dataset
credit_card_data.info()
```

```python
# check for missing values
credit_card_data.isnull().sum()
```

```python
#check distribution of classes
credit_card_data['Class'].value_counts()
```

Dataset is higly unbalanced
0 --> normal transaction
1 --> fraudulent transcation

```python
#separate data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

Mean is higher on fraudulent transactions

```python
# compare values for both classes
credit_card_data.groupby('Class').mean()
```

```python
#Under-Sampling
#Build a Sample Dataset containing similiar distribution of both classes
legit_sample = legit.sample(n=492)
```

```python
#Concactenate two data frames
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
# Split data into features and targets
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
#Split data into training data and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
#Train machine learning model (Logistic Regression)
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
#Evaluate the model (Accuracy score)
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data:', training_data_accuracy)
```

```python
#Evaluate with Test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test data:', test_data_accuracy)
```
